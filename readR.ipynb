```python
import pyreadr
import pandas as pd
import csv
```

```python
result = pyreadr.read_r('subredsimdata.rdata')
```

```python
df_similarity = pd.DataFrame(result['subsimmat'])
```

```python
df_similarity.sample(10)
```

```python
# make pivot on df
df_pivot = df_similarity.stack().reset_index()
df_pivot.columns = ['SOURCE', 'TARGET', 'SIMILARITY']

df_pivot.sample(10)
```

```python
# to csv
df_pivot.to_csv('subreddit_similarity.csv', index=False)
```
